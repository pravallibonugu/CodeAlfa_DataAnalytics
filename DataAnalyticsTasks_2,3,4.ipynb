```python
import re

def clean_name(name):
    if pd.isna(name):
        return name
    # Split on \r\n and take unique parts, then rejoin
    parts = [p.strip() for p in re.split(r'\r\n', str(name))]
    unique_parts = list(dict.fromkeys(parts))  # removes duplicates, keeps order
    return ' / '.join(unique_parts) if len(unique_parts) > 1 else unique_parts[0]

df['name_clean'] = df['name'].apply(clean_name)

df['name_clean'].value_counts().head(10)
```

```python
# Create a new column: review length (word count)
df['review_length'] = df['reviews.text'].apply(lambda x: len(str(x).split()))

df[['reviews.rating', 'review_length']].describe()
```

```python
plt.figure(figsize=(9,6))
sns.boxplot(x='reviews.rating', y='review_length', data=df, hue='reviews.rating', palette='coolwarm', legend=False)
plt.title('Review Length vs Rating')
plt.xlabel('Rating')
plt.ylabel('Review Length (words)')
plt.ylim(0, 200)  # zoom in, since a few very long reviews can skew the scale
plt.show()
```

```python
# Convert date column to datetime
df['reviews.date'] = pd.to_datetime(df['reviews.date'], errors='coerce')

# Extract year-month
df['year_month'] = df['reviews.date'].dt.to_period('M')

# Average rating per month
monthly_rating = df.groupby('year_month')['reviews.rating'].mean()

plt.figure(figsize=(12,5))
monthly_rating.plot(kind='line', marker='o', color='purple')
plt.title('Average Rating Over Time')
plt.xlabel('Month')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
review_counts = df.groupby('year_month').size()

plt.figure(figsize=(12,4))
review_counts.plot(kind='bar', color='gray')
plt.title('Number of Reviews per Month')
plt.xlabel('Month')
plt.ylabel('Review Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filter to the well-populated range only
reliable_range = df[(df['year_month'] >= '2015-11') & (df['year_month'] <= '2017-09')]

monthly_rating_reliable = reliable_range.groupby('year_month')['reviews.rating'].mean()

plt.figure(figsize=(12,5))
monthly_rating_reliable.plot(kind='line', marker='o', color='purple')
plt.title('Average Rating Over Time (Nov 2015 – Sep 2017, reliable range)')
plt.xlabel('Month')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
!pip install textblob
from textblob import TextBlob

sample_text = df['reviews.text'].iloc[0]
print(sample_text)

blob = TextBlob(sample_text)
print(blob.sentiment)
```

```python
def get_sentiment(text):
    polarity = TextBlob(str(text)).sentiment.polarity
    if polarity > 0.1:
        return 'Positive'
    elif polarity < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment'] = df['reviews.text'].apply(get_sentiment)

df['sentiment'].value_counts()
```

```python
# Cross-tab: sentiment vs actual rating
pd.crosstab(df['reviews.rating'], df['sentiment'])
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='reviews.rating', hue='sentiment', data=df, palette='Set2')
plt.title('Sentiment Label vs Star Rating')
plt.xlabel('Star Rating')
plt.ylabel('Number of Reviews')
plt.legend(title='Sentiment')
plt.show()
```

```python
# Find 1-star reviews that were labeled "Positive"
misclassified = df[(df['reviews.rating'] == 1.0) & (df['sentiment'] == 'Positive')]

# Show a few examples
for text in misclassified['reviews.text'].head(5):
    print(text)
    print('---')
```
